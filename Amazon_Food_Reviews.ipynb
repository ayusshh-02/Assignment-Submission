```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
df=df[['Score','Summary','Text']]
```

```python
df
```

```python
# we are adding a new column name Positivity which will work as a label for our model which will show whether a review is positive or negative
# if Score> 3 then it is positive and if Score<3 review is negative and if score==3 review is neutral.
# Our model is of binary Classification so we need not to include reviews with rating 3 in our model.

df=df[df['Score']!=3]
df['Positivity']=np.where(df['Score']>3,1,0)
```

```python
df
```

```python
# Now Text column will work as feature and Positivity Column will work as label .
feature=df['Text']
label=df['Positivity']
```

```python
feature
```

```python
label
```

```python
# now our Text column data isin NLP form so we need to vectorized it. Before vectorization we have to perform data cleanup activity on feature data.
cleaned_review=[]
for review in feature:
    review= re.sub('[^A-Za-z]',' ',review)
    review= review.lower()
    review= review.split()
    review= [word for word in review if word not in stopwords.words('english')]
    ps=PorterStemmer()
    review= [ps.stem(word) for word in review]
    review= " ".join(review)
    cleaned_review.append(review)
```

```python
cleaned_review 
```

```python
cleaned_feature= cleaned_review
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer(min_df=5).fit(cleaned_feature)
```

```python
len(vectorizer.get_feature_names())
```

```python
vectorized_features= vectorizer.transform(cleaned_feature)
```

```python
vectorized_features
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, labels_train, labels_test = train_test_split(vectorized_features,label,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(features_train,labels_train)
```

```python
pred=model.predict(features_test)
pred
```

```python
# finding score of the model
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(labels_test,pred)
```

```python
# accuracy score of the model
print('Accuracy score of the model is {}%'.format(accuracy_score(labels_test,pred)*100))
```

```python
no_of_checks=int(input("Enter no of checks:"))
for i in range(no_of_checks):
  review=input("enter review:")
  check_feature=vectorizer.transform([review])
  prediction=model.predict(check_feature)
  if prediction==1:
    print("{} : This Review is Positive".format(review))
  else:
    print("{} : This Review is Negative".format(review))  
```
